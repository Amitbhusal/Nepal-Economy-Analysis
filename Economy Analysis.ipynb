## Consumption Analysis

### Consumption And Investment 

```python
sns.set(rc={"figure.figsize":(15, 5)})

# Melt the DataFrame to have "variable" and "value" columns
df_melted = df.melt(id_vars=["FiscalYear(AD)"], value_vars=["Investment", "Consumption"], 
                    var_name="Category", value_name="Amount")

sns.barplot(data=df_melted, x="FiscalYear(AD)", y="Amount", hue="Category")

# Adding labels and title
plt.xlabel("Fiscal Year")
plt.ylabel("Amount")
plt.title("Comparison of Investment and Consumption for Each Fiscal Year")
plt.xticks(rotation=45)
plt.legend(title="Category")
plt.show()
```

#### When every year the investment is much lower than consumption, it generally indicates that the country is relying heavily on spending and consumption without proportionate investment in production and growth. This can lead to imbalances, such as reduced savings, limited capacity for future growth, and potentially a less robust economy over time. In essence, it suggests a potential risk of economic instability or slowdown due to the mismatch between consumption and investment.

```python
print(df)
```

### Gross Domestic and  National  Savings 


```python
df["GDP_Deflator"] = (df["GDP (Current Price)"] / df["GDP (Constant Price)"]) * 100
#plt.figure(figsize=(10, 6))
#plt.figure(facecolor='lightgrey')
plt.plot(df["Start Year"],df["GDP_Deflator"],marker = "*",ls ="--",color ="Green",)
plt.xlabel("Fiscal Year")
plt.ylabel("GDP  Deflator")
plt.title("GDP Deflator over time")
```

```python
sns.lineplot(data = df,x ="Start Year",y = "Gross Domestic Savings",marker = "*",color = "red")
```

#### GDS is increasing at a slower rate than the GDP Deflator, it suggests that the growth in savings is not keeping up with the rise in prices, which may impact future consumption and investment.

```python
sns.lineplot(data = df,x ="Start Year",y = "Gross National Savings",marker = "*",color = "brown")
```

#### GNS is increasing faster than the GDP Deflator, it suggests that the residents' total savings, including foreign savings, are growing at a faster rate than the overall price level

#### If both are happening at the same time ,it means the country reiles more on foreign investor and people within the country are not investing in the proper rate and sending their money abroad .

### ** Comparision of Nominal GNI and Real GNI ** 

```python
df["Real GNI"] = (df["Gross National Income"]/df["GDP_Deflator"])*100
#print(df["Real GNI"])
plt.plot(df["Start Year"],df["Real GNI"],marker = "*",label = "Real GNI",color = "red")
plt.plot(df["Start Year"],df["Gross National Income"],marker = "*",label = "Nominal GNI",color = "Green")
plt.legend()
plt.show()
```

 ####  Real GNI lags behind nominal GNI, it suggests that inflation is eroding the actual purchasing power of income.
#### The increase in nominal income is not enough to compensate for the increase in prices due to inflation.
